```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("water.csv")
df.head()
```

```python
fig,ax=plt.subplots()
df['height'].plot(ax=ax)
ax.set_title("Example")
ax.axhline(0,linestyle='--', color='green')
```

```python
#y=x^3
#y2=13x^2-y-1000
x=range(-10,11)
y=np.power(x,3)

fig,ax=plt.subplots()
ax.plot(x,y,'gd')
fig.savefig("demo.png")
ax.set_title("Third power of x")
```

```python
fig,ax=plt.subplots()
y2=13*np.power(x,2)-y-1000
ax.plot(x,y2,'rx')
ax.set_title("just a polnominal")
```

```python
x=range(-10,11)
y=np.power(x,3)
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y,'gd')
fig.savefig("demo.png")
ax[0].set_title("Third power of y")
```
