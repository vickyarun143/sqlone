```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
df=pd.read_csv('sales.csv',parse_dates=['Date'])
df['Month']=df['Date'].dt.month
sales=df.groupby(['Month','Product'])['Amount'].sum().unstack()
sales
```

```python
fig,ax=plt.subplots()
sales.plot(ax=ax)
```

```python
styles={'a':':s','b':'--v','c':'-o'}

with plt.style.context('seaborn-dark'):
    fig,ax=plt.subplots(figsize=(6,3))
    for column in styles:
        sales[column].dropna().plot(ax=ax,style=styles)
        
```

```python
fig.suptitle("sales2017",fontweight='bold')
ax.set_title("monthly sales for products a,b,c",fontstyle='italic',fontsize=10)
```
