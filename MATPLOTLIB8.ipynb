```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
```

```python
#MAIN PROPERTIES OF AN AXIS 
#LIMITS:VALUE WHERE THE AXIS STARTS AND ENDS
#TICKS:POSITIONS OF INDICATOR LINES I=ON THE AXIS
#TICK LABELS:TEXT SHOWN AT TICKS
```

```python
df=pd.read_csv('sales.csv',parse_dates=['Date'])
df['Month']=df['Date'].dt.month
sales=df.groupby(['Month','Product'])['Amount'].sum().unstack()
sales.head()
```

```python
styles={'a':':s','b':'--v','c':'-o'}

with plt.style.context('ggplot'):
    fig,ax=plt.subplots(figsize=(6,3))
    for column in styles:
        sales[column].dropna().plot(ax=ax,style=styles[column],grid=True,ylim=(0,2000),legend=True,label='Product'+column)
ax.set_title("Monthly sales")
ax.set_ylabel('sales(EUR)')
```

```python
ax.set_xtricks(range(1,13))
```
